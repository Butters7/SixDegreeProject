Теория шести рукопожатий — социологическая теория, согласно которой любые два человека на Земле разделены не более чем пятью уровнями общих знакомых (и, соответственно, шестью уровнями связей). Формальная математическая формулировка теории — диаметр графа знакомств не превышает 6.

Теория шести рукопожатий является интересным феноменом, который впервые был сформулирован в 1929 году венгерским писателем Фридрихом Каринти. Он выдвинул гипотезу о том, что каждый человек на планете может быть связан с другим любым человеком через цепочку знакомых длиной не более шести уровней. Однако, именно в 1967 году социолог Стэнли Милгрэм провел знаменитый эксперимент "Малый мир", который стал основой для дальнейших исследований и развития этой теории.

Эксперимент Милгрэма состоял в том, чтобы выбрать случайного жителя Великобритании и передать ему письмо с просьбой доставить его определенному человеку, о котором изначально знали только его имя и место жительства. Для передачи письма участникам эксперимента разрешалось отправлять его только знакомым, которые могли быть связаны с искомым получателем через своих собственных знакомых. Результаты эксперимента показали, что в среднем письма достигали целевых лиц, используя примерно шесть промежуточных знакомых.

Давайте рассмотрим следующий пример, чтобы продемонстрировать верность этой теории:

Предположим, что у нас есть два человека: Алиса и Боб, которые живут в разных странах и никогда не встречались друг с другом. Цель состоит в том, чтобы установить связь между Алисой и Бобом через цепочку знакомых.

Шаг 1: Алиса знает Дэйва, своего коллегу по работе.

Шаг 2: Дэйв знаком с Клер, старым другом, с которым он учился в университете.

Шаг 3: Клер знакома с Эриком, своим соседом по дому.

Шаг 4: Эрик знаком с Фрэнком, своим тренером по футбольной команде.

Шаг 5: Фрэнк знаком с Грейс, которая является сотрудницей местной библиотеки.

Шаг 6: Грейс знакома с Бобом, так как он постоянно посещает библиотеку.

Таким образом, мы построили цепочку знакомств, состоящую из шести человек, связывающую Алису и Боба:

Алиса -> Дэйв -> Клер -> Эрик -> Фрэнк -> Грейс -> Боб

Этот пример подтверждает верность теории шести рукопожатий. Хотя Алиса и Боб могут быть географически удалены друг от друга и никогда не встречались, они все равно могут быть связаны через цепочку знакомств, не превышающую шести человек.

Маркетинг и реклама: Использование теории шести рукопожатий может помочь в определении влиятельных личностей или целевых групп, а также в разработке стратегий маркетинга и рекламы, основанных на социальных связях и взаимодействии между людьми.
